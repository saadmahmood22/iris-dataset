```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

# Reading csv file into pandas dataframe
iris = pd.read_csv("modified_iris.csv")

# Printing out the first few rows of the dataframe
iris.head(5)
```

```python
# Scatterplot of petal length vs petal width

plt.style.use("seaborn")

PL = iris["Petal_length(mm)"]
PW = iris["Petal_width(mm)"]

plt.figure(figsize = (8, 5))
plt.scatter(PL, PW)

plt.xlabel("Petal Length (mm)")
plt.ylabel("Petal Width (mm)")

plt.show()
```

```python
# There appears to be a linear relationship between petal length and petal width.
# When petal length increases, petal width also increases.
```

```python
# Scatterplot of sepal length vs sepal width

plt.style.use("seaborn")

SL = iris["Sepal_length(mm)"]
SW = iris["Sepal_width(mm)"]

plt.figure(figsize = (8, 5))
plt.scatter(SL, SW)

plt.xlabel("Sepal Length (mm)")
plt.ylabel("Sepal Width (mm)")

plt.show()
```

```python
# The scatterplot of sepal length vs sepal width does not indicate a linear relationship between the two variables.
# Sepal length and sepal width are almost independent of eachother.
```

```python
# Scatterplot of petal length vs sepal length

plt.style.use("seaborn")

PL = iris["Petal_length(mm)"]
SL = iris["Sepal_length(mm)"]

plt.figure(figsize = (8, 5))
plt.scatter(PL, SL)

plt.xlabel("Petal Length (mm)")
plt.ylabel("Sepal Length (mm)")

plt.show()
```

```python
# There is a roughly linear relationship between petal length and sepal length.
```

```python
# Scatterplot of petal width vs sepal width

plt.style.use("seaborn")

PW = iris["Petal_width(mm)"]
SW = iris["Sepal_width(mm)"]

plt.figure(figsize = (8, 5))
plt.scatter(PW, SW)

plt.xlabel("Petal Width (mm)")
plt.ylabel("Sepal Width (mm)")

plt.show()
```

```python
# There does not appear to be a linear relationship between petal width and sepal width
```

```python
# Scatterplot of petal length vs sepal width

plt.style.use("seaborn")

PL = iris["Petal_length(mm)"]
SW = iris["Sepal_width(mm)"]

plt.figure(figsize = (8, 5))
plt.scatter(PL, SW)

plt.xlabel("Petal Length (mm)")
plt.ylabel("Sepal Width (mm)")

plt.show()
```

```python
# Scatterplot of sepal length vs petal width

plt.style.use("seaborn")

SL = iris["Sepal_length(mm)"]
PW = iris["Petal_width(mm)"]

plt.figure(figsize = (8, 5))
plt.scatter(SL, PW)

plt.xlabel("Sepal Length (mm)")
plt.ylabel("Petal Width (mm)")

plt.show()
```

```python
# Plotting histograms which use different colours for each type of Species.

iris_setosa = iris.loc[iris["Species"]=="setosa"]
iris_versicolor = iris.loc[iris["Species"]=="versicolor"]
iris_virginica = iris.loc[iris["Species"]=="virginica"]

sb.FacetGrid(iris,hue="Species",height=3).map(sb.histplot,"Petal_length(mm)").add_legend()
sb.FacetGrid(iris,hue="Species",height=3).map(sb.histplot,"Petal_width(mm)").add_legend()
sb.FacetGrid(iris,hue="Species",height=3).map(sb.histplot,"Sepal_length(mm)").add_legend()
sb.FacetGrid(iris,hue="Species",height=3).map(sb.histplot,"Sepal_width(mm)").add_legend()
plt.show()
```

```python
# The plots above show that we can identify and separate flowers belonging to the iris-setosa Species
# based on petal length and petal width.

# Sepal length and Sepal width do not provide us with any information since the data is very scattered
# and we cannot separate the flowers.
```

```python
plt.style.use("default")

plt.figure(figsize = (5, 7))

setosa_PL = iris.loc[iris.Species == "setosa"]["Petal_length(mm)"]
versicolor_PL = iris.loc[iris.Species == "versicolor"]["Petal_length(mm)"]
virginica_PL = iris.loc[iris.Species == "virginica"]["Petal_length(mm)"]

lb = ["setosa", "versicolor", "virginica"]

plt.boxplot([setosa_PL, versicolor_PL, virginica_PL], labels = lb)

plt.ylabel("Petal Length (mm)")

plt.show()
```

```python
# The boxplot above also highlights the fact that we can identify Iris-Setosa flowers based on Petal Length
# since the length of their petals is completely unique

# Iris-setosa has the most outliers. The sknewness of all three datasets is not that great.
```

```python
plt.style.use("default")

plt.figure(figsize = (5, 7))

setosa_PW = iris.loc[iris.Species == "setosa"]["Petal_width(mm)"]
versicolor_PW = iris.loc[iris.Species == "versicolor"]["Petal_width(mm)"]
virginica_PW = iris.loc[iris.Species == "virginica"]["Petal_width(mm)"]

lb = ["setosa", "versicolor", "virginica"]

plt.boxplot([setosa_PW, versicolor_PW, virginica_PW], labels = lb)

plt.ylabel("Petal Width (mm)")

plt.show()
```

```python
# All three datasets are positively skewed.

# The petal length of Iris-setosa is extremely skewed in the positive direction. The median appears 
# to be equal to the lower quartile.

# Iris-versicolor and Iris-virginica do not have any outliers.

# Petal width can also be used to identify Iris-Setosa.
```

```python
plt.style.use("default")

plt.figure(figsize = (5, 7))

setosa_SL = iris.loc[iris.Species == "setosa"]["Sepal_length(mm)"]
versicolor_SL = iris.loc[iris.Species == "versicolor"]["Sepal_length(mm)"]
virginica_SL = iris.loc[iris.Species == "virginica"]["Sepal_length(mm)"]

lb = ["setosa", "versicolor", "virginica"]

plt.boxplot([setosa_SL, versicolor_SL, virginica_SL], labels = lb)

plt.ylabel("Sepal Length (mm)")

plt.show()
```

```python
# The boxplot comparison shows that it is very difficult to identify the flower based on
# sepal length since the values are similar.

# On average, setosa flowers have the smallest sepal length, followed by versicolor and virginica
# flowers have the largest sepal length. This can be seen by the median values (orange line)
```

```python
plt.style.use("default")

plt.figure(figsize = (5, 7))

setosa_SW = iris.loc[iris.Species == "setosa"]["Sepal_width(mm)"]
versicolor_SW = iris.loc[iris.Species == "versicolor"]["Sepal_width(mm)"]
virginica_SW = iris.loc[iris.Species == "virginica"]["Sepal_width(mm)"]

lb = ["setosa", "versicolor", "virginica"]

plt.boxplot([setosa_SW, versicolor_SW, virginica_SW], labels = lb)

plt.ylabel("Sepal Width (mm)")

plt.show()
```

```python
# Sepal width is randomly scattered, and the boxplot comparison does not provide us with valuable information.
```
